### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
# Dataframe Information
print('--- Dataframe Information ---')
df.info()

# Dataframe Summaries
print('--- Dataframe Summaries ---')
print(df.describe(include='all'))

# Dataframe Shape
print('--- Dataframe Shape ---')
print(f'Rows: {df.shape[0]} \n Columns: {df.shape[1]}')
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Dataset Missing Values
print('--- Dataset Missing Values ---')
print(df.isnull().sum())
```

```python
# Create a list of columns with missing values
missingvalue_cols = ['RemoteWork', 'JobSatPoints_11', 'SurveyLength', 'SurveyEase', 'ConvertedCompYearly', 'JobSat']

# Calculate percentage of missing values
missing_info = df[missingvalue_cols].isnull().mean() * 100
print("Percentage of missing values per column:")
print(missing_info)
```

```python
# Fill missing RemoteWork with the most frequent value
remotework_mode = df['RemoteWork'].mode()[0]
df['RemoteWork'] = df['RemoteWork'].fillna(remotework_mode)

# Fill missing SurveyLength with the most frequent value
surveylength_mode = df['SurveyLength'].mode()[0]
df['SurveyLength'] = df['SurveyLength'].fillna(surveylength_mode)

# Fill missing SurveyEase with the most frequent value
surveyease_mode = df['SurveyEase'].mode()[0]
df['SurveyEase'] = df['SurveyEase'].fillna(surveyease_mode)

# Fill missing ConvertedCompYearly with the median value
salary_median = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(salary_median)

# Drop rows where JobSat is missing
df = df.dropna(subset=['JobSat'])

# Verify missing values
print('--- Cleaned Dataset Missing Values ---')
print(df.isnull().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
# # Distribution of Employment, JobSat, YearsCodePro Status
print(f'--- Employment value count:--- \n{df['Employment'].value_counts()}')
print(f'--- JobSat value count: --- \n{df['JobSat'].value_counts()}')
print(f'--- YearsCodePro value count: --- \n{df['YearsCodePro'].value_counts()}')
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
# Calculate the frequency of each unique JobSat score
jobsat_counts = df['JobSat'].value_counts()

# Plot the pie chart
jobsat_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Job Satisfaction Distribution')
plt.ylabel('') 
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# Process 'Worked With' languages
worked_with = df['LanguageHaveWorkedWith'].str.split(';').explode().value_counts()

# Process 'Want to Work With' languages
want_to_work = df['LanguageWantToWorkWith'].str.split(';').explode().value_counts()

# Combine both into a single DataFrame
comparison_df = pd.concat([worked_with, want_to_work], axis=1)
comparison_df.columns = ['WorkedWith', 'WantToWorkWith']

# Select the top 10 languages
top_10_comparison = comparison_df.sort_values(by='WorkedWith', ascending=False).head(10)

# Reset index to turn languages into a column for plotting
top_10_comparison = top_10_comparison.reset_index().rename(columns={'index': 'Language'})

# Convert columns to rows for grouped bars
melted_df = top_10_comparison.melt(id_vars='Language', var_name='Type', value_name='Count')

# Create a Grouped Bar Chart
plt.figure(figsize=(10, 5))
sns.barplot(data=melted_df, x='Language', y='Count')

plt.title('Current Usage vs. Desired Future Usage (Top 10 Languages)')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)

plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
# Identify the Top 10 countries by number of respondents
top_10_countries = df['Country'].value_counts().nlargest(10).index

# Filter the dataframe to only include these countries
df_top_regions = df[df['Country'].isin(top_10_countries)]

# Create the Grouped Bar Chart
plt.figure(figsize=(15, 5))

# Create Grouped Bar Chart
sns.countplot(
    data=df_top_regions, 
    x='Country', 
    hue='RemoteWork', 
    order=top_10_countries
)

plt.title('Remote Work Distribution by Country (Top 10)')
plt.xlabel('Country / Region')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
# Clean YearsCodePro to make it numeric
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0, 
    'More than 50 years': 50
})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop rows where either JobSat or YearsCodePro has missing values
correlation_data = df[['JobSat', 'YearsCodePro']].dropna()

# Calculate Pearson Correlation
pearson_corr = correlation_data['JobSat'].corr(correlation_data['YearsCodePro'], method='pearson')

print(f"Pearson Correlation: {pearson_corr:.4f}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Shorten EdLevel values
short_ed = {
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master',
    'Some college/university study without earning a degree': 'Some College',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'High School',
    'Other doctoral degree (Ph.D., Ed.D., etc.)': 'PhD'
}
df['EdLevel_Short'] = df['EdLevel'].map(short_ed)

# Group Employment into categories
emp_mapping = {
    'Employed, full-time': 'Employed',
    'Employed, part-time': 'Employed',
    'Independent contractor, freelancer, or self-employed': 'Self-Employed',
    'Not employed, but looking for work': 'Not Employed',
    'Not employed, and not looking for work': 'Not Employed',
    'Retired': 'Not Employed',
    'I prefer not to say': 'Other',
    'Student, full-time': 'Student',
    'Student, part-time': 'Student'
}
df['Employment_Grouped'] = df['Employment'].map(emp_mapping)

# Filter for Top 5 EdLevels and drop NA rows
top_ed = df['EdLevel_Short'].value_counts().nlargest(5).index
df_clean = df[df['EdLevel_Short'].isin(top_ed)].dropna(subset=['Employment_Grouped'])

# Create normalised cross tabulation
ct_grouped = pd.crosstab(df_clean['EdLevel_Short'], 
                         df_clean['Employment_Grouped'], 
                         normalize='index')

# Create stacked bar plot visualisation
ax = ct_grouped.plot(kind='bar', 
                     stacked=True, 
                     figsize=(10, 5),
                    )

plt.title('Grouped Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Percentage of Group')
plt.legend(title='Employment Category', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
